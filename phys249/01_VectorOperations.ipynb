### <font color=#0072DA> Vectors in $\mathbb{R}^2$, $\mathbb{R}^3$, and $\mathbb{R}^n$</font>
#### <font color=#003599>Key Concepts </font>
Based on the Cartesian coordinate system for $\mathbb{R}^2$ and $\mathbb{R}^3$, we're used to seeing ordered pairs $(x,y)$ or triples $(x,y,z)$ as coordinates for a singular point in 2-space or 3-space. When we think of these in an abstract manner, we convert a point $P(p_1,p_2)$ to:
$$
\vec{p} =
\begin{bmatrix}
p_1 \\
p_2
\end{bmatrix}
$$

<div class="alert alert-block alert-info"> &#x1F4A1; <b>DEFINITION</b> Let $\mathbb{R}^2$ denote the set of all vectors of the form $\begin{bmatrix} x_1 \\ x_2 \end{bmatrix}$ where $x_1$ and $x_2$ are real numbers called the <b>components</b> of a vector. Then, $$\mathbb{R}^2 = \left\{ \begin{bmatrix} x_1 \\ x_2 \end{bmatrix} \mid x_1, x_2 \in \mathbb{R} \right\}$$</div>

<div class="alert alert-block alert-info"> &#x1F4A1; <b>DEFINITION</b> Let $\vec{x} = \begin{bmatrix} x_1 \\ x_2 \end{bmatrix}; \vec{y} = \begin{bmatrix} y_1 \\ y_2 \end{bmatrix} \in \mathbb{R}^2$. Then we define the following, where it is understood that $\vec{x}-\vec{y}$ is interpreted as $\vec{x}+(-1)\vec{y}$:<br>
$$\text{Vector Addition:} \quad \vec{x} + \vec{y} = \begin{bmatrix} x_1 \\ x_2 \end{bmatrix} + \begin{bmatrix} y_1 \\ y_2 \end{bmatrix} = \begin{bmatrix} x_1 + y_1 \\ x_2 + y_2 \end{bmatrix}$$
$$\text{Scalar Multiplication:} \quad t\vec{x} = t\begin{bmatrix} x_1 \\ x_2 \end{bmatrix} = \begin{bmatrix} tx_1 \\ tx_2 \end{bmatrix}$$</div>

**linear combinations:** the sums of scalar multiples

The **vector equation of a line** is given by: $\vec{x} = \vec{p}+t\vec{d} \mid t \in \mathbb{R}$, where $\vec{d}$ is the **direction vector** and $\vec{p}$ is the point that the line passes through. We commonly break this down into its components, or **parametric equations**, given by:
$$\left\{ \begin{matrix} x_1 = p_1 + td_1 \\ x_2 = p_2 + td_2 \end{matrix}, \quad t\in \mathbb{R} \right.$$

The **vector equation of a plane** through a point $\vec{p}$ in $\mathbb{R}^3$ is defined as: $\vec{x} = \vec{p} + t\vec{u} + s\vec{v}, \quad s,t \in \mathbb{R}$. 
Keep in mind that if the vectors $\vec{u}$ or $\vec{v}$ are scalar multiples of each other, then $\vec{x}$ is not a plane but rather a line!

```python
# =============================================================================
# INTRODUCING `QUIVER` FUNCTION IN MATPLOTLIB
# =============================================================================

## 1. set an origin and create a vector in 2D space
myOrigin = np.array ([0,0]) 
myVector = np.array ([5,1])

## 2. call on quiver function to plot the vector
plt.quiver(*myOrigin, *myVector, color='b', angles='xy', scale_units='xy', scale=1)

### `*` -> grabbing the individual values and plotting them

## 3. add an x and y limit, plot graph
plt.xlim(0,10)
plt.ylim(0,10)
plt.show()
```

```python
# =============================================================================
# PLOTTING 2 VECTORS ON A PLOT
# =============================================================================

## 1. set an origin and create 2 vectors in 2D space
myOrigins = np.array ([[0,0], [0,0]])
myVectors = np.array ([[5,1], [4,8]])

## 2. call on quiver function
plt.quiver(*myOrigins, myVectors[:,0], myVectors[:,1], color=['b','g'], angles='xy',scale_units='xy', scale=1)

### plt.quiver(xStartValues,yStartValues, EndValues, yEndValues, extra parameters)

### `*` -> grabbing the individual values and plotting them

plt.xlim(0,10)
plt.ylim(0,10)
plt.show()
```

```python
# =============================================================================
# ADDING VECTORS AND CALCULATING RESULTANT
# =============================================================================

## 1. create resultant vector by adding the 2 vectors
resultant = myVectors[0] + myVectors[1]

## 2. call on `quiver` fucntion for resultant and individual vectors
plt.quiver(*myOrigins, *resultant, color='r', angles='xy', scale_units='xy', scale=1)
plt.quiver(*myOrigins, myVectors[:,0], myVectors[:,1], color=['b','g'], angles='xy',scale_units='xy', scale=1)

## 3. add x and y limits
plt.xlim(0,10)
plt.ylim(0,10)

## 4. label the vectors with text
plt.text(myVectors[0,0] + 0.1, myVectors[0,1] + 0.1, 'Vector 1', fontsize=12, color='b')
plt.text(myVectors[1,0] + 0.1, myVectors[1,1] + 0.1, 'Vector 2', fontsize=12, color='g')
plt.text(resultant[0] + 0.1, resultant[1] + 0.1, 'Resultant', fontsize=12, color='r')

## 5. plot graph
plt.show()
print(f"The resultant is given as {resultant}")
```

```python
# =============================================================================
# PLOTTING VECTOR EQUATIONS OF LINES AND PLANES
# =============================================================================

## 1. define the starting point as the position vector (p) as well as the direction vector (d)
p = np.array([2,3])
d = np.array([3,1])

### the equation of a line is `x = p + td`
### therefore, find a bunch of vectors with different t values

## 2. generate values of t to get points along the line using np.linspace
tValues = np.linspace(-10, 10, 100)  ## -> splitting up -10 to 10 into 100 values, can adjust range as needed

### np.linspace (start of range, end of range, number of values)

## 3. calculate points on the line using vector equation x(t) = p + t*d
linePoints = np.array([p + t*d for t in tValues]) ## -> this is in vector form

## 4. set up and plot the graph
plt.figure(figsize=(4,4))
plt.plot(linePoints[:,0], linePoints[:,1], label="Vector Eqn of Line", color="darkorchid", zorder=1)
plt.quiver(p[0], p[1], d[0], d[1], angles='xy', scale_units='xy', scale=1,color='darkgreen', label="Direction Vector d", zorder=2) ### -> z order - order of stacking, higher number gets a higher layer
plt.scatter(*p, color='cyan', zorder=3, label="Starting Point p")
plt.xlim(-20,20)
plt.ylim(-20,20)
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)
plt.grid()
plt.legend()
plt.show()
```

```python
# =============================================================================
# PLOTTING PLANES (3D GRAPHS)
# =============================================================================

## 1. define the known point on the plane (d) and the two direction vectors (u, v)
p = np.array([2, 3, 1])   ### -> a point on the plane
u = np.array([3, 1, 2])   ### -> first direction vector
v = np.array([0, 3, 1])   ### -> second direction vector

## 2. generate a grid of values for s and t to cover the plane, adjust the range and density of the values as needed
tValues = np.linspace(-10, 10, 10)
sValues = np.linspace(-10, 10, 10)
t, s = np.meshgrid(tValues, sValues)

## 3. calculate points on the plane using the vector equation x(u, v) = p + t*u + s*v
planePoints = p[:, np.newaxis, np.newaxis] + t[np.newaxis, :, :] * u[:, np.newaxis, np.newaxis] + s[np.newaxis, :, :] * v[:, np.newaxis, np.newaxis]

## 4. extract X, Y, and Z coordinates from plane_points
X, Y, Z = planePoints[0], planePoints[1], planePoints[2]

## 5. set up and plot - much easier to use fig and ax when using 3D plots
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, color='darkorchid', alpha=0.2, edgecolor='darkgreen')

## 6. plot the starting point r0 and direction vectors v1 and v2
ax.quiver(*p, *u, color='blue', label='Direction Vector u')
ax.quiver(*p, *v, color='fuchsia', label='Direction Vector v')
ax.scatter(*p, color='red', s=50, label='Point p')

## 7. add labels and grid 
ax.set_title('Vector Equation of a Plane')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.legend()
plt.show()
```

### <font color=#0072DA>Dot Products, Cross Products and Projections</font>
#### <font color=#003599>Key Concepts </font>

<div class="alert alert-block alert-info"> &#x1F4A1; <b>DEFINITION</b> The <b>norm</b> or <b>length</b> of a vector $\| \vec{x} \|$ is given by the following equation if $\vec{x}$ exists in $\mathbb{R}^n$:
$$\|\vec{x}\| = \sqrt{x_1^2 + x_2^2 + \dots + x_n^n}$$</div>

<div class="alert alert-block alert-info"> &#x1F4A1; <b>DEFINITION</b> The <b>dot product</b> is given by:
$$\vec{x} \cdot \vec{y} = \|\vec{x}\|\|\vec{y}\| \cos\theta, \quad \vec{x} \cdot \vec{y} = x_1 y_1 + x_2 y_2 + \dots + x_n y_n$$
And the length can also be defined as $\|\vec{x}\| = \sqrt{\vec{x}\cdot \vec{x}}$.
We note that two vectors are <b>orthogonal</b> to each other if and only if $\vec{x} \cdot \vec{y} = 0$.</div>

<div class="alert alert-block alert-info"> &#x1F4A1; <b>DEFINITION</b> The <b>cross product</b> is given by:
$$\vec{x} \times \vec{y} = \|\vec{x}\|\|\vec{y}\| \sin\theta$$
However, to do this manually, one must calculate it as follows: $$\vec{x} \times \vec{y} = \begin{bmatrix} x_2 y_3 - x_3 y_2 \\ x_3 y_1 - x_1 y_2 \\ x_1 y_2 - x_2 y_1 \end{bmatrix}$$</div>

Meanwhile, the **projection** of $\vec{y}$ onto $\vec{x}$ is given by the following equation. In other words, we are trying to find out how much of the vector $\vec{y}$ is along the direction of $\vec{x}$:
$$\text{proj}_\vec{x}(\vec{y}) = \frac{\vec{x}\cdot\vec{y}}{\|\vec{x}\|^2}\vec{x}$$

```python
# =============================================================================
# FINDING THE DOT PRODUCT WITH NUMPY
# =============================================================================

### introduction to np.dot()

## 1. define 2 vectors
a = np.array([-4,3,2])
b = np.array([9,5,2])

## 2. find dot product using np.dot()
dotProduct = np.dot(a,b)
print(f"The dot product is {dotProduct}")
```

### <font color=#0072DA>6.3 Physics Applications</font>

**SCENARIO:** A force is applied to move an object along a certain path. Calculate the work done by the force, by projecting the force vector onto the direction of displacement. The work done by the force is then given by the dot product of the force vector and the displacement vector, so vector projections are needed.

```python
# =============================================================================
# DEFINED FUNCTIONS FOR VECTOR PROJECTION AND WORK DONE
# =============================================================================
# import numpy as np

def vectorProjection(a,b):
    """
    Calculate the projection of vector a onto vector b
    Param: a (vector) and b (direction vector a will be projected onto)
    Return: the projection of a onto b in vector form
    """
    
    ## ensure that the vectors a and b are NumPy arrays
    a = np.array(a)
    b = np.array(b)
    
    ## get the numerator of the vector projection (dot product of `a` and `b`)
    dotProduct_ab = np.dot(a,b)
    
    ## get the denominator of the vector projection (dot product of `b` and `b`)
    dotProduct_bb = np.dot(b,b)
    
    ## calculate vector projection
    projection = (dotProduct_ab)/(dotProduct_bb) * b
    
    return projection

def workDone(force, displacement):
    """
    Calculate the work done by a force along a given displacement
    Param: vector force (vector), vector displacement (displacement)
    Return: work in units of J
    """
    
    ## project the force vector onto the displacement vector
    forceProjection = vectorProjection(force, displacement)
    
    ## work done is the magnitude of the projection of force in the direction of displacement
    work = np.linalg.norm(forceProjection)
    
    return work


### test the functions with some values

qState = np.array([1,2]) ### -> `a` vector
bState = np.array([3,4]) ### -> `b` vector
print(f"The project of {qState} onto {bState} is: {vectorProjection(qState,bState)}")

force = np.array([10, 15, 20])
displacement = np.array([25, 30, 5])
print(f"Work done by force {force} N, along {displacement} m, is {workDone(force, displacement)} J")
```
